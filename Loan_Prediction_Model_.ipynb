## **Introduction**
A loan refers to a sum of money that is borrowed by an individual or organization from a lender, with the agreement to repay the borrowed amount along with interest within a specified period. The goal of loan prediction modeling is to assess the creditworthiness of the applicants and predict whether they are likely to repay the loan based on various factors such as credit history, applicantincome, loan amount, loan term, and other relevant features.


## **Problem Statement**
XYZ bank aims to automate the loan eligibility process in real-time, based on customer details provided in online application forms. With factors such as "Gender", "Married", "Education", "Dependents", "ApplicantIncome", "Loan_Amount", "Credit History", and others taken into consideration, the goal is to identify eligible customer segments for specific loan amounts. By automating this process, the bank can efficiently target these eligible customers, streamlining the loan approval process and reducing manual errors in applicant selection.
It is a classification problem where we have to predict whether a loan would be approved or not.

After exploring all the variables in our data, we will impute the missing values and treat the outliers. Missing data and outliers can adversely affect the model performance.






#### Handling Missing Values

For numerical variables: imputation using mean or median

For categorical variables: imputation using mode

```python
# check for actual number of missing values in train data
df_train.isnull().sum()
```

There are very few missing values in the Gender, Married, Dependents, Credit_History, and Self_Employed features. We can fill them using the mode of the respective features.

```python
df_train['Gender'].fillna(df_train['Gender'].mode()[0], inplace=True)
df_train['Married'].fillna(df_train['Married'].mode()[0], inplace=True)
df_train['Dependents'].fillna(df_train['Dependents'].mode()[0], inplace=True)
df_train['Self_Employed'].fillna(df_train['Self_Employed'].mode()[0], inplace=True)
df_train['Credit_History'].fillna(df_train['Credit_History'].mode()[0], inplace=True)
```

```python
df_train.isnull().sum()
```

Loan_Amount_Term:

```python
#checking the Loan_Amount_Term  count
df_train['Loan_Amount_Term'].value_counts()
```

The value of 360 is the most common , so we will fill the missing values with the mode

```python
# filling the missing values with the mode
df_train['Loan_Amount_Term'].fillna(df_train['Loan_Amount_Term'].mode()[0], inplace=True)
```

 LoanAmount:

 Since "LoanAmount" is a numerical variable and we have outliers in the data, it would be more appropriate to use the median to impute the missing values. The mean can be highly affected by the presence of outliers, so using the median is a more robust approach in this case.






```python
#filling the missing value with the median
df_train['LoanAmount'].fillna(df_train['LoanAmount'].median(), inplace=True)
```

### Feature Selection

Target variable (Loan_Status) will be seperated from Predictor variables.

Sklearn requires the target variable in a separate dataset. So, we will drop our target variable from the training dataset

Create feature matrix X and target vector y. Where our target is "Loan_Status", and our features are the columns that remain in the encoded dataframe.

```python
df_train_encode.columns
```

```python
features=['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Credit_History',  'Gender',
       'Married', 'Dependents_1', 'Dependents_2', 'Dependents_4', 'Education',
       'Self_Employed', 'Property_Area_Semiurban', 'Property_Area_Urban']
target= "Loan_Status"
```

### Handling Data Inbalance

```python
y.value_counts()
```

It can be observed that there is an imbalance between the number of approved and rejected loans. To avoid bias towards approved loans, oversampling techniques will be used.

```python
# Display the class distribution before SMOTE
print("Class distribution before SMOTE:", Counter(y))
```

```python
# Apply SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
X, y = smote.fit_resample(X, y)
```

```python
# Display the class distribution after SMOTE
print("Class distribution after SMOTE:", Counter(y))
```

The distribution of loan statuses is now balanced.

#### Model Improvement with Hyperparemeter tuning:

We will try to improve the accuracy by tuning the hyperparameters for this model. We will use grid search to get the optimized values of hyperparameters. Grid-search is a way to select the best of a family of hyperparameters, parametrized by a grid of parameters.

```python
from sklearn.model_selection import GridSearchCV
```

```python
paramgrid = {'max_depth': list(range(1, 20, 2)),
                    'n_estimators': list(range(1, 200, 20))}
gridSearchModel=GridSearchCV(RandomForestClassifier(random_state=1),paramgrid)
```

```python
# fit the grid search model
gridSearchModel.fit(X_train, y_train)
```

```python
# Estimating the optimized value
gridSearchModel.best_estimator_
```

The optimized value for the max_depth variable is 19 and for n_estimator is 81. We will build the model using these optimized values.

```python
RFC_grid = RandomForestClassifier(max_depth=19, n_estimators=81,random_state=1)
CvScores_rfc_grid= cross_val_score(RFC_grid, X,y, cv=kfold)
for i, score in enumerate(CvScores_rfc_grid, start=1):
    print(f"Fold {i}: {score:.2f}")
print(f"Average Accuracy: {CvScores_rfc_grid.mean():.2f}")
```
